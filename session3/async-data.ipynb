 Async Data Pattern

```typescript
const state = {
    initial: 'initial',
    loading: 'loading',
    loaded: 'loaded',
    error: 'error'
} as const
```

```typescript
type State = typeof state[keyof typeof state]
```

```typescript
type AsyncData<T> = {
    state: State
    data?: T
    error?: string
}
```
