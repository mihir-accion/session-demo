 Create Types based on objects

```typescript
const state = {
    initial: 'initial',
    loading: 'loading',
    loaded: 'loaded',
    error: 'error'
} as const
```

```typescript
type State = typeof state[keyof typeof state]
```

```typescript
type LoadedResult<T> = {
    state: 'loaded'
    data: T
}

type ErrorResult = {
    state: 'error'
    error: string
}
```

```typescript
type Result<T> = { state: 'initial' | 'loading' } | LoadedResult<T> | ErrorResult
```
