# Utility Types

```python
// Consider User database entity
type UserEntity = {
    id: number
    firstname: string
    middlename?: string
    lastname: string
    age: number
    email: string
}
```

### 1. Partial

```python
type UserPartial = Partial<UserEntity>
```

### 2. Omit

```python
type UserDto = Omit<UserEntity, 'id'>
```

### 3. Required

```python
type User = Required<UserEntity>
```
