# Type Operators

And and Or Operator

```typescript
type Value<T> = { data: T } | { error: string }

type User = { name: string } & { age: number }

const value1: Value<string> = { data: 'data' }
const user: User = {
    name: 'Jon skeet',
    age: 40
}
```

**typeof** operator

```typescript
const point = {
    x: 50,
    y: 100
}

type Point = typeof point
```

```typescript
const MyArray = [
  { name: "Alice", age: 15 },
  { name: "Bob", age: 23 },
  { name: "Eve", age: 38 },
];
 
type Person = typeof MyArray[number];

type Age = typeof MyArray[number]["age"];
```

**keyof** operator 

```typescript
type P = keyof Point;
```

Use Cases: Helpful in creating enums

```typescript
// In code, we need to use proper name but value should be in numbers

const orderStatus = {
    pending: 1,
    delivered: 2,
    returned: 3
} as const

// orderStatus.pending = 400;

type OrderStatus = typeof orderStatus[keyof typeof orderStatus]

function hasDelivered(status: OrderStatus): boolean {
    return status !== 1;
}
```

Ternary Operator

```typescript
type Result<T, U> = T extends string ? U : T;

function parseValue<T, U>(input: T): Result<T, U> {
  if (typeof input === 'string') {
    // Type assertion because TS can't infer runtime type narrowing for conditional types
    return JSON.parse(input) as Result<T, U>;
  }
  return input as Result<T, U>;
}

// Usage:
const value = parseValue<string, Point>(`{"x": 100, "y": 10}`);
const value2 = parseValue<number, Point>(5);
```
