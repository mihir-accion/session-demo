Pub Sub Pattern

```typescript
import { Subject, map } from 'npm:rxjs';

let i = 1
const observable = new Subject<number>();

const id = setInterval(() => {
    observable.next(i++);
}, 2000)

observable.subscribe((value) => console.log('Subscriber 1:', value));

observable
.pipe(map(p => p * 2))
.subscribe((value) => console.log('Subscriber 2:', value));

setTimeout(() => {
    clearInterval(id);
}, 10000)

console.log("After subscriber")
```
