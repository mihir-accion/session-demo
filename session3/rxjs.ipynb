Pub Sub Pattern

```typescript
import { Subject } from 'npm:rxjs';

let i = 1
const observable = new Subject<number>();

const id = setInterval(() => {
    observable.next(i++);
}, 2000)

observable.subscribe({
  next: (value) => console.log('Received:', value),
  complete: () => console.log('Done!')
});

setTimeout(() => {
    clearInterval(id);
}, 10000)

console.log("After subscriber")
```
