```python
// Consider User database entity
type UserEntity = {
    id: number
    firstname: string
    middlename?: string
    lastname: string
    age: number
    email: string
}

type Order = {
    id: number
    product: {
        id: number
        name: string
    },
    status: 'pending' | 'success' | 'failed'
}
```

```python
type UserWithOrder = Omit<UserEntity, 'id'> & { 
    orders: Omit<Order, 'id'>[]
}
```

Make types more readable

```python
type Prettify<T> = {
    [K in keyof T]: T[K]
}
```

```python
type Prettify_UserWithOrder = Prettify<UserWithOrder>
```

Custom Omit type

```python
type StrictOmit<T, K extends keyof T> = Omit<T, K>
```

```python
type MyOmit<T, K extends keyof T> = {
    [P in keyof T as P extends K ? never : P]: T[K]
}
```

Complex Type Manipulation

```python
type TypeGetter<T> = {
    [K in keyof T as `get${Capitalize<string & K>}`]-?: () => T[K]
}
```

```python
type UserGetter = TypeGetter<UserEntity>
```
