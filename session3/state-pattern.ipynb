# State Pattern

Implement Async data as state

```python
type AsyncData<T> = { state: 'initial' }
                    | { state: 'loading' }
                    | { state: 'loaded', data: T }
                    | { state: 'error', error: string }
```

```python
type User = {
    name: string,
    age: number
}

const user: AsyncData<User> = {
    state: 'loaded',
    data: {
        name: 'John',
        age: 40
    }
}
```

```python
if(user.state === 'loaded') {
    user.data
}
```

Helper function

```python
function getData<T>(asyncData: AsyncData<T>) {
    return asyncData.state === 'loaded' ? asyncData.data : undefined;
}
```
